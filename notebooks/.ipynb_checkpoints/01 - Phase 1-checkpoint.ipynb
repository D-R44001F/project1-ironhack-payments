## Relevant Lexique Details

### **Fees Dataset**

#### Key Columns:
- **id:** Unique ID for the fee.
- **type:** Type of fee (instant payment, incident, etc.).
- **status:** Status of the fee (accepted, rejected, etc.).
- **category:** Reason for the incident.
- **reason:** Description of the fee.

---

### **Cash Requests Dataset**

#### Key Columns:
- **id:** Unique ID for the cash request.
- **amount:** Amount requested.
- **status:** Status of the request (rejected, accepted, etc.).
- **created_at:** Date and time when the request was created.
- **user_id:** Unique user identification.
- **reimbursement_date:** Planned reimbursement date.
- **cash_request_received_date:** Date the request was received.
- **money_back_date:** Date the money was considered returned.
